# Group by 

```python
df_deliveries_new['total_runs_y'] = pd.to_numeric(df_deliveries_new['total_runs_y'], errors='coerce')
```

```python
df_deliveries_new.info()
```

```python
# compute per-match cumulative runs (select the numeric column before cumsum)
df_deliveries_new['cumulative_runs'] = df_deliveries_new.groupby('match_id')['total_runs_y'].cumsum()
```

```python
# replace NaNs (if any) and cast to int for convenience
df_deliveries_new['cumulative_runs'] = df_deliveries_new['cumulative_runs'].fillna(0).astype(int)
df_deliveries_new
```

```python
df_deliveries_new['runs_left'] = df_deliveries_new['total_runs_x'] - df_deliveries_new['cumulative_runs']
# df_deliveries_new runs left is total runs in inning 1 - cumulative runs in inning 2
```

```python
df_deliveries_new
```

```python
df_deliveries_new['balls_left'] = 126 - (df_deliveries_new['over']*6 + df_deliveries_new['ball'])
```

# Find Wickets

```python
# compute wickets remaining per delivery using df_deliveries_new
# create a binary indicator (1 if a player was dismissed on that delivery, else 0)
df_deliveries_new['player_dismissed_ind'] = df_deliveries_new['player_dismissed'].notna().astype(int)

# cumulative wickets taken per match
df_deliveries_new['wickets_taken'] = df_deliveries_new.groupby('match_id')['player_dismissed_ind'].cumsum()

# wickets remaining (assuming 10 wickets per innings)
df_deliveries_new['wickets_left'] = 10 - df_deliveries_new['wickets_taken']
```

```python
# show relevant columns for verification
df_deliveries_new[['player_dismissed','player_dismissed_ind','wickets_taken','wickets_left']].tail(20)
```

# Finding Current Runrate

```python
# crr = runs/overs
(df_deliveries_new['cumulative_runs']*6) / (df_deliveries_new['over']*6 + df_deliveries_new['balls_left'])
```

```python
df_deliveries_new['current_runrate'] = (df_deliveries_new['cumulative_runs']*6) / (df_deliveries_new['over']*6 + df_deliveries_new['balls_left'])
```

```python
# Finding Required Run Rate (rrr)
```

```python
# rrr = runs/overs (required run rate)
(df_deliveries_new['runs_left']*6)/df_deliveries_new['balls_left']
```

```python
df_deliveries_new['required_runrate'] = (df_deliveries_new['runs_left']*6)/df_deliveries_new['balls_left']
```

# RESULT 

```python
def result(row):
    return 1 if row['batting_team'] == row['winner'] else 0
```

```python
df_deliveries_new.apply(result, axis=1)
```

```python
df_deliveries_new['result'] = df_deliveries_new.apply(result, axis=1)
```

# SPLIT TRAINING AND TEST DATA

```python
from sklearn.model_selection import train_test_split
```

```python
X = df_final.iloc[:,:-1] # all columns except the last one and is features as X
y = df_final.iloc[:,-1] # last column and is target variable as y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.2, random_state=1)
```

```python
X_train
```

# APPLY ONE HOT ENCODING FOR CATEGORICAL / STRING COLUMNS

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
trf = ColumnTransformer(
    [
        ('trf', OneHotEncoder(sparse_output=False, drop='first'), ['batting_team','bowling_team','city'])
    ],
    remainder='passthrough'
)
```

```python
# Model train using LogisticRegression
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
```

```python
pipe = Pipeline(steps=[
    ('step1', trf),
    ('step2', LogisticRegression(solver='liblinear'))
])
```

```python
pipe.fit(X_train, y_train)
```

```python
y_pred = pipe.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

```python
pipe.predict_proba(X_test)[10]
```

# Model TRAIN USING SVC

```python
from sklearn.svm import SVC
from sklearn.pipeline import Pipeline
# use the existing transformer `trf` to encode categorical columns before SVC
pipe_svc = Pipeline([('trf', trf), ('svc', SVC())])
pipe_svc.fit(X_train, y_train)
pipe_svc.score(X_test, y_test)
```

```python
pipe_svc.predict(X_test)[10]
```

# Model TRAIN USING DecisionTreeClassifier

```python
from sklearn.tree import DecisionTreeClassifier
# pipeline with transformer to encode categorical features

pipe_dt = Pipeline([('trf', trf), ('dt', DecisionTreeClassifier())])
pipe_dt.fit(X_train, y_train)
pipe_dt.score(X_test, y_test)
```

```python
pipe_dt.predict_proba(X_test)[10]
```

# Model TRAIN USING RandomForestClassifier

```python
from sklearn.ensemble import RandomForestClassifier
# pipeline with transformer to encode categorical features
pipe_rf = Pipeline([('trf', trf), ('rf', RandomForestClassifier(n_estimators=250))])
pipe_rf.fit(X_train, y_train)
pipe_rf.score(X_test, y_test)
```

```python
pipe_rf.predict_proba(X_test)[10]
```
